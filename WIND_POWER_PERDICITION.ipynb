```python
from flask import Flask, request, jsonify
import pandas as pd
import pickle # Import pickle to load the model

# Initialize the Flask app
app = Flask(__name__)

# Load the trained model within the Flask app's context
try:
    with open('model2.pkl', 'rb') as file:
        loaded_model = pickle.load(file)
    print("Model 'model2.pkl' loaded successfully within Flask app.")
except Exception as e:
    print(f"Error loading model: {e}")
    loaded_model = None # Set to None if loading fails

# Define the root endpoint for a simple health check
@app.route('/', methods=['GET'])
def home():
    return "Turbine ActivePower Prediction API is running!"

# Define the prediction endpoint
@app.route('/predict', methods=['POST'])
def predict():
    if loaded_model is None:
        return jsonify({'error': 'Machine learning model not loaded.'}), 500
    try:
        # Get JSON data from the request
        data = request.get_json(force=True)

        # Extract features expected by the model
        # Ensure all required features are present and are integers
        required_features = ['hour', 'minute', 'day', 'month', 'year', 'dayofweek', 'dayofyear']
        input_features = {}
        for feature in required_features:
            if feature not in data or not isinstance(data[feature], int):
                return jsonify({'error': f'Missing or invalid feature: {feature}. Expected integer.'}), 400
            input_features[feature] = [data[feature]]

        # Create a DataFrame from the input features
        # The model was trained on a DataFrame with these columns
        features_df = pd.DataFrame(input_features)

        # Make prediction using the loaded model
        prediction = loaded_model.predict(features_df)[0]

        # Return the prediction as a JSON response
        return jsonify({'prediction': prediction.item()}) # .item() to convert numpy float to standard Python float

    except Exception as e:
        return jsonify({'error': str(e)}), 500

print("Flask application and /predict endpoint defined.")
```

```python
from flask import Flask, request, jsonify
import pandas as pd
import pickle # Import pickle to load the model

# Initialize the Flask app
app = Flask(__name__)

# Load the trained model within the Flask app's context
try:
    with open('model2.pkl', 'rb') as file:
        loaded_model = pickle.load(file)
    print("Model 'model2.pkl' loaded successfully within Flask app.")
except Exception as e:
    print(f"Error loading model: {e}")
    loaded_model = None # Set to None if loading fails

# Define the root endpoint for a simple health check
@app.route('/', methods=['GET'])
def home():
    return "Turbine ActivePower Prediction API is running!"

# Define the prediction endpoint, now without API key authentication
@app.route('/predict', methods=['POST'])
def predict():
    if loaded_model is None:
        return jsonify({'error': 'Machine learning model not loaded.'}), 500
    try:
        # Get JSON data from the request
        data = request.get_json(force=True)

        # Extract features expected by the model
        required_features = ['hour', 'minute', 'day', 'month', 'year', 'dayofweek', 'dayofyear']
        input_features = {}
        for feature in required_features:
            if feature not in data or not isinstance(data[feature], int):
                return jsonify({'error': f'Missing or invalid feature: {feature}. Expected integer.'}), 400
            input_features[feature] = [data[feature]]

        features_df = pd.DataFrame(input_features)

        prediction = loaded_model.predict(features_df)[0]

        return jsonify({'prediction': prediction.item()}) # .item() to convert numpy float to standard Python float

    except Exception as e:
        return jsonify({'error': str(e)}), 500

print("Flask application and /predict endpoint redefined without API key authentication.")
```

```python
import threading
import time
import socket

FLASK_PORT = 5000
FLASK_BASE_URL = f"http://127.0.0.1:{FLASK_PORT}"

# Global flag to track if our Flask thread was started
# Initialize it if it doesn't exist to handle the first run
if 'flask_app_thread_active' not in globals():
    flask_app_thread_active = False
# Also initialize flask_server_thread to None if it doesn't exist
if 'flask_server_thread' not in globals():
    flask_server_thread = None

# Function to run the Flask app
def run_flask_app_target(port):
    global app, flask_app_thread_active
    try:
        app.run(debug=True, use_reloader=False, host='0.0.0.0', port=port)
        # This line is typically not reached as app.run() blocks
        flask_app_thread_active = False # If it somehow exits, set flag to False
    except Exception as e:
        print(f"Error starting Flask app on port {port}: {e}")
        flask_app_thread_active = False

# Attempt to start the Flask app if it's not currently active,
# or if the existing thread has unexpectedly died.
if not flask_app_thread_active or (flask_server_thread and not flask_server_thread.is_alive()):
    print(f"Attempting to start Flask app on port {FLASK_PORT} in a new thread...")
    # Create and start the thread
    flask_server_thread = threading.Thread(target=run_flask_app_target, args=(FLASK_PORT,))
    flask_server_thread.daemon = True # Allows main program to exit even if thread is still running
    flask_server_thread.start()
    time.sleep(3) # Give Flask a moment to bind to the port

    if flask_server_thread.is_alive():
        print(f"Flask app thread successfully started on port {FLASK_PORT}.")
        flask_app_thread_active = True
    else:
        print("Flask app thread failed to start unexpectedly. This might be due to the port being in use or an error within the Flask application itself.")
        flask_app_thread_active = False
elif flask_app_thread_active and flask_server_thread and flask_server_thread.is_alive():
    print("Flask app thread is already running.")
else:
    print("Flask app thread state is ambiguous. Attempting to force a restart on next execution.")
    flask_app_thread_active = False # Reset flag to allow restart on next run

# Define the local prediction URL
PREDICT_URL_LOCAL = f"{FLASK_BASE_URL}/predict"

print(f" * Flask app running locally at: {FLASK_BASE_URL}")
print(f" * Prediction endpoint available locally at: {PREDICT_URL_LOCAL}")
print("Note: This API is not publicly exposed and can only be accessed within this Colab environment.")
```

```python
import ipywidgets as widgets
from IPython.display import display
import calendar
from datetime import date

# 2. Create IntText widget for 'hour'
hour_input = widgets.IntText(
    value=10,
    min=0,
    max=23,
    description='Hour (0-23):',
    disabled=False
)

# 3. Create IntText widget for 'minute'
minute_input = widgets.IntText(
    value=30,
    min=0,
    max=59,
    description='Minute (0-59):',
    disabled=False
)

# 4. Create IntText widget for 'day'
day_input = widgets.IntText(
    value=15,
    min=1,
    max=31,
    description='Day (1-31):',
    disabled=False
)

# 5. Create IntText widget for 'month'
month_input = widgets.IntText(
    value=5,
    min=1,
    max=12,
    description='Month (1-12):',
    disabled=False
)

# 6. Create IntText widget for 'year'
year_input = widgets.IntText(
    value=2023,
    min=2017, # Assuming data starts from 2017 as per original dataset
    max=2025, # Arbitrary upper limit for future predictions
    description='Year (YYYY):',
    disabled=False
)

# 7. Create Dropdown widget for 'dayofweek'
dayofweek_input = widgets.Dropdown(
    options={'Monday': 0, 'Tuesday': 1, 'Wednesday': 2, 'Thursday': 3, 'Friday': 4, 'Saturday': 5, 'Sunday': 6},
    value=3, # Default to Wednesday
    description='Day of Week:',
    disabled=False,
)

# 8. Create IntText widget for 'dayofyear'
dayofyear_input = widgets.IntText(
    value=135, # Default for May 15
    min=1,
    max=366,
    description='Day of Year (1-366):',
    disabled=False
)

# 9. Create a Button widget
predict_button = widgets.Button(description='Get Prediction')

# 10. Create an Output widget to display results
output_area = widgets.Output()
validation_output = widgets.Output() # New output for validation messages

def validate_date(*args):
    with validation_output:
        validation_output.clear_output()
        try:
            year = year_input.value
            month = month_input.value
            day = day_input.value

            if not (1 <= month <= 12):
                print("Invalid month. Please enter a value between 1 and 12.")
                predict_button.disabled = True
                return

            if not (1 <= day <= calendar.monthrange(year, month)[1]):
                print(f"Invalid day for month {month} in year {year}. Day must be between 1 and {calendar.monthrange(year, month)[1]}.")
                predict_button.disabled = True
                return

            d = date(year, month, day)
            dayofweek_input.value = d.weekday()
            dayofyear_input.value = d.timetuple().tm_yday

            # print("Date input valid.") # Optional: print success message
            predict_button.disabled = False # Enable button if date is valid
        except ValueError as e:
            print(f"Date validation error: {e}. Please check year, month, and day inputs.")
            predict_button.disabled = True
        except Exception as e:
            print(f"An unexpected error occurred during date validation: {e}")
            predict_button.disabled = True

# Attach validation function to changes in day, month, year inputs
day_input.observe(validate_date, names='value')
month_input.observe(validate_date, names='value')
year_input.observe(validate_date, names='value')

# Trigger initial validation to set correct dayofweek/dayofyear and button state
validate_date()

print("ipywidgets for input and output have been created with real-time validation.")
```

```python
import requests
import json

def on_predict_button_clicked_validated(b):
    with output_area:
        output_area.clear_output()
        # The validate_date function is now automatically called on input changes,
        # and disables the button if input is invalid.
        # So, if the button is enabled, we assume the date is valid.

        # Re-run validation one last time to ensure `dayofweek` and `dayofyear` are up-to-date
        # and to catch any very last-second invalid states if not caught by observers.
        # This call also enables/disables the button.
        validate_date()

        if predict_button.disabled:
            print("Cannot predict: Please correct invalid date inputs.")
            return

        try:
            # 2. Access values from input widgets
            input_data = {
                'hour': hour_input.value,
                'minute': minute_input.value,
                'day': day_input.value,
                'month': month_input.value,
                'year': year_input.value,
                'dayofweek': dayofweek_input.value, # These are updated by validate_date
                'dayofyear': dayofyear_input.value  # These are updated by validate_date
            }

            print(f"Sending data to API: {input_data}")

            # 5. Send HTTP POST request to the local Flask /predict endpoint
            headers = {'Content-Type': 'application/json'}
            response = requests.post(PREDICT_URL_LOCAL, data=json.dumps(input_data), headers=headers)

            # 6. Implement error handling
            if response.status_code == 200:
                prediction_result = response.json()
                # 7. Process and print the prediction result
                print(f"Prediction: {prediction_result['prediction']:.2f}")
            else:
                print(f"API request failed with status code {response.status_code}")
                print(f"Error response: {response.json()}")

        except requests.exceptions.ConnectionError as e:
            print(f"Connection error: {e}. Make sure the Flask app is running locally on port {FLASK_PORT}.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

# 8. Link the function to the predict button's on_click event
predict_button.on_click(on_predict_button_clicked_validated)

print("Prediction logic function defined and linked to the button with pre-prediction validation.")
```
